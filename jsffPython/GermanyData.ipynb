```python
#录入数据：X、Y存原数据，x、y1、y2存拟合数据
X = np.linspace(0,120,121)
Y = np.array(df2)
Y = Y.reshape(-1)
x = np.linspace(0.001,125,12500)
y1 = -1786.77 + 133.823*x  -0.970403*x*x
y2 = np.exp(-11.416 + 0.55978*x + -0.00389091*x*x)
#坐标轴
ax = plt.gca() 
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.spines['left'].set_color('black') 
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
#绘制曲线、点及图例
l1, = plt.plot(x,y1,color='b',linewidth=1.0,linestyle='-')
l2, = plt.plot(x,y2,color='r',linewidth=1.0,linestyle='-')
p1 = plt.scatter(X,Y,s=20,color='green')
plt.legend(handles = [l1,l2,p1], labels = ["y = -1786.77 + 133.823x + -0.970403x^2",
                                           "y = e^(-11.416 + 0.55978x + -0.00389091x^2)",
                                           "OriginalData"], loc = 4,fontsize = 7)
#绘制图像
plt.show()
```
