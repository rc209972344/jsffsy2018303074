```python
import matplotlib.pyplot as plt
import numpy as np
#录入数据：X、Y存原数据，x、y1、y2存拟合数据
X = np.linspace(1,16,16)
Y = np.array([0.004,0.0064,0.008,0.0088,0.00922,0.0095,0.0097,0.00986,0.010,
     0.0102,0.01032,0.01042,0.0105,0.01055,0.01058,0.0106])
x = np.linspace(0.001,25,2500)
y1 = x/(80.1745*x + 162.723)
y2 = 0.0113252*np.exp(-1.05668/x)
#坐标轴
ax = plt.gca() 
ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none') 
ax.spines['left'].set_color('black') 
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))
#绘制曲线、点及图例
l1, = plt.plot(x,y1,color='b',linewidth=1.0,linestyle='-')
l2, = plt.plot(x,y2,color='r',linewidth=1.0,linestyle='-')
p1 = plt.scatter(X,Y,s=20,color='green')
plt.legend(handles = [l1,l2,p1], labels = ["y=x/(80.1745x+162.723)",
                                           "y=0.0113252e^(-1.05668/x)",
                                           "OriginalData"], loc = 7,fontsize = 10)
#绘制图像
plt.show()
```
